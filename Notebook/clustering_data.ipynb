```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler= StandardScaler()
```

```python
df[['RGOB_T','NPHI_T','GR_T','PEF_T','DTC_T']] = scaler.fit_transform(df[['RHOB','GR','NPHI','PEF','DTC']])
```

```python
df
```

```python
def optimise_K_means (data,max_k):
    means=[]
    inertias=[]
    for k in range(1,max_k):
        kmeans= KMeans(n_clusters=k)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)
    fig=plt.subplots(figsize=(10,5))
    plt.plot(means, inertias,'o-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```

# Function for find optimised number of clusters

```python
optimise_K_means(df[['RGOB_T','NPHI_T']],10)
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(df[['RGOB_T','NPHI_T']])
```

```python
df['KMeans_5']= kmeans.labels_
```

```python
plt.scatter(x=df['NPHI'], y=df['RHOB'], c=df['KMeans_5'])
plt.xlim(-0.1,1)
plt.ylim(3,1.5)
plt.show()
```

```python
for k in range(1,6):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(df[['RGOB_T','NPHI_T']])
    df[f'KMeans_{k}'] = kmeans.labels_
```
